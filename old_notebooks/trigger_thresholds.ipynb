```python
import pandas as pd
import numpy as np
import runDB
import matplotlib.pyplot as plt
from analyze import get_thresholds

%matplotlib inline
```

```python
collection = runDB.get_collection()
query = {'source.type' : {'$ne' : 'LED'},
         '$and' : [{'number' : {'$gt' : 6820}},
                   {'number' : {'$lt' : 6890}}
                  ]
        }
cursor = collection.find(query, {'number' : 1})

runlist = sorted([r['number'] for r in cursor])
#runlist = np.array([run['number'] for run in cursor 
#                 if any([ r['register'] == '8060' for r in run['reader']['ini']['registers']])])

def compare_thresholds(list1, list2):
    if len(list1) != len(list2):
        return True, None
    pre_mask = [l1==l2 for l1,l2 in zip(list1, list2)]
    mask = [not i for i in pre_mask]
    changes = np.where(mask)[0]
    val_changes = [(ch, list1[ch], list2[ch]) for ch in changes]
    return any(mask), val_changes


previous = []
for r in runlist:
    thresholds = get_thresholds(r)
    did_change, changes = compare_thresholds(previous, thresholds)
    if did_change:
        print("Threshold change at run %d" % r)
        if changes:
            for change in changes:
                print('\tCh %d: %d --> %d' % (change))
    previous = thresholds
    
    
```

```python
old_thresholds = get_thresholds(6821)
new_thresholds = get_thresholds(6846)
np.savez('selftrigger_thresholds_SR1_before6846', thresholds=old_thresholds)
np.savez('selftrigger_thresholds_SR1_after6846', thresholds=new_thresholds)
```
